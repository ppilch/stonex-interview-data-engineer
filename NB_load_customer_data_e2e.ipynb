```python
#fetching data from src ff
src_path=f"{lakehouse_path}{file_path}"
print(src_path)

df_customers = spark.read.option("header", "true").csv(src_path)

#adding index column and etl timestamp
df_customers=df_customers.withColumn("Index", monotonically_increasing_id())\
                         .withColumn("ETL_POST_TS", current_timestamp())

print(df_customers.columns)

df_customers=df_customers.select(
    "Index"
    ,'CustomerId'
    , 'FirstName'
    , 'LastName'
    , 'Company'
    , 'City'
    , 'Country'
    , 'Phone1'
    , 'Phone2'
    , 'Email'
    , 'SubscriptionDate'
    , 'Website'
    , 'ETL_POST_TS'
)

display(df_customers)

#saving bronze
bronze_table_path = f"{lakehouse_path}/Tables/bronze_customer"
df_customers.write.format("delta").mode("overwrite").option("overwriteSchema", "True").save(bronze_table_path)
```

```python
# read data from bronze
df_bronze_customer_data = spark.read.format("delta").load(bronze_table_path)

#data validation - filter out records when  where both FirstName and LastName are missing.
df_bronze_customer_data = df_bronze_customer_data.filter(
    (col("FirstName").isNotNull()) & (col("FirstName") != "") &
    (col("LastName").isNotNull()) & (col("LastName") != "")
)

#read country data 
dim_country_path = f"{lakehouse_path}/Tables/gold_dimcountry"
df_gold_dim_country = spark.read.format("delta").load(dim_country_path)
df_gold_dim_country.select("CountryKey","CountryName")


# display(df_gold_dim_country)
df_bronze_customer_data = df_bronze_customer_data.join(df_gold_dim_country, df_bronze_customer_data["Country"] == df_gold_dim_country["CountryName"], "left")

#transform data
df_bronze_customer_data=df_bronze_customer_data.withColumn("CustomerName",concat(col("FirstName"),lit(" "),col("LastName")))\
                                               .withColumn("Phone1",expr("CASE WHEN Phone1 RLIKE '^[0-9\\+\\(]' THEN Phone1 ELSE 'Unknown' END"))\
                                               .withColumn("Phone2",expr("CASE WHEN Phone2 RLIKE '^[0-9\\+\\(]' THEN Phone2 ELSE 'Unknown' END"))\
                                               .withColumn("EmailDomain",split(col("Email"), "@")[1])\
                                               .withColumn("SubscriptionDate",to_date(col("SubscriptionDate"), "dd/MM/yyyy"))\
                                               .withColumn("SubscriptionDateKey", date_format(col("SubscriptionDate"), "yyyyMMdd").cast("int"))


#selection of columns for silver 
df_bronze_customer_data=df_bronze_customer_data.select(
    "Index"
    ,'CustomerId'
    , 'CustomerName'
    , 'Company'
    , 'CountryKey'
    , 'Country'
    , 'Phone1'
    , 'Phone2'
    , 'Email'
    ,'EmailDomain'
    , 'SubscriptionDateKey'
)


#add sentinels 
df_bronze_customer_data=df_bronze_customer_data.fillna({
    "Index":-1,
    'CustomerId':-1,
    'CustomerName':"Unknown",
    'Company':"Unknown",
    'CountryKey':-1,
    'Country':"Unknown",
    'Phone1':"Unknown",
    'Phone2':"Unknown",
    'Email':"Unknown",
    'EmailDomain':"Unknown",
    'SubscriptionDateKey':-1
})


display(df_bronze_customer_data)

#saving silver
silver_table_path = f"{lakehouse_path}/Tables/silver_customer"
df_bronze_customer_data.write.format("delta").mode("overwrite").option("overwriteSchema", "True").save(silver_table_path)
```
